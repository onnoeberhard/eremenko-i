# Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Importing the dataset

```python
dataset = pd.read_csv('Data.csv')
X = dataset.iloc[:, :-1].values    # All columns but the last
y = dataset.iloc[:, 3].values      # just the last column

dataset.style.background_gradient()    # warning because of NaN values.
```

# Taking care of missing data

```python
from sklearn.preprocessing import Imputer
imputer = Imputer(missing_values = 'NaN', strategy = 'mean', axis = 0)
imputer = imputer.fit(X[:, 1:3])
X[:, 1:3] = imputer.transform(X[:, 1:3])
```

# Encoding categorical data

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

# Encoding the Independent Variable
labelencoder_X = LabelEncoder()
X[:, 0] = labelencoder_X.fit_transform(X[:, 0])
onehotencoder = OneHotEncoder(categorical_features = [0])     # encoding categorical data into dummy variables
X = onehotencoder.fit_transform(X).toarray()

# Encoding the Dependent Variable
labelencoder_y = LabelEncoder()
y = labelencoder_y.fit_transform(y)

display(pd.DataFrame(X).style.set_caption('X').background_gradient())
display(pd.DataFrame(y).style.set_caption('y').background_gradient())
```

# Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
sc_y = StandardScaler()
# y_train = sc_y.fit_transform(y_train) -- Unnecessary here b/c y only has 0 or 1 as value.

display(pd.DataFrame(X_train).style.set_caption('X Train').background_gradient())
display(pd.DataFrame(y_train).style.set_caption('y Train').background_gradient())
display(pd.DataFrame(X_test).style.set_caption('X Test').background_gradient())
display(pd.DataFrame(y_test).style.set_caption('y Test').background_gradient())
```
