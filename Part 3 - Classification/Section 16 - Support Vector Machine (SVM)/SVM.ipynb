```python
# Importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
# Importing the dataset
dataset = pd.read_csv("Social_Network_Ads.csv")
dataset
```

```python
X = dataset.iloc[:, [2, 3]].values
y = dataset.Purchased.values
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/4)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

```python
# Fitting SVM to the training set
from sklearn.svm import SVC
classifier = SVC(kernel='linear', probability=True)    # probability=True enables `predict_proba` function
classifier.fit(X_train, y_train)
```

```python
# Predicting the test set results
y_pred = classifier.predict(X_test)
y_pred
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
# Visualising the training set results
X_set, y_set = X_train, y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 0.5, stop = X_set[:, 0].max() + 0.5, step = 1e-2),
                     np.arange(start = X_set[:, 1].min() - 0.5, stop = X_set[:, 1].max() + 0.5, step = 1e-2))
plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape), alpha=0.9)
plt.scatter(X_set[:, 0], X_set[:, 1], c=y_set, edgecolors="black")
plt.title('SVM (Training set)')
plt.legend((plt.Line2D((0,1),(0,0), marker='o', color=matplotlib.cm.get_cmap()(0.), markeredgecolor='black', linestyle=''),
            plt.Line2D((0,1),(0,0), marker='o', color=matplotlib.cm.get_cmap()(1.), markeredgecolor='black', linestyle='')), 
           ("No", "Yes"))
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.show()
```

```python
# Visualising the test set results
X_set, y_set = X_test, y_test
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 0.5, stop = X_set[:, 0].max() + 0.5, step = 1e-2),
                     np.arange(start = X_set[:, 1].min() - 0.5, stop = X_set[:, 1].max() + 0.5, step = 1e-2))
plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape), alpha=0.9)
plt.scatter(X_set[:, 0], X_set[:, 1], c=y_set, edgecolors="black")
plt.title('SVM (Test set)')
plt.legend((plt.Line2D((0,1),(0,0), marker='o', color=matplotlib.cm.get_cmap()(0.), markeredgecolor='black', linestyle=''),
            plt.Line2D((0,1),(0,0), marker='o', color=matplotlib.cm.get_cmap()(1.), markeredgecolor='black', linestyle='')), 
           ("No", "Yes"))
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.show()
```

```python
# Visualising the test set results (probabilities)
X_set, y_set = X_test, y_test
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 0.5, stop = X_set[:, 0].max() + 0.5, step=1e-3),
                     np.arange(start = X_set[:, 1].min() - 0.5, stop = X_set[:, 1].max() + 0.5, step=1e-3))
plt.contourf(X1, X2, classifier.predict_proba(np.array([X1.ravel(), X2.ravel()]).T)[:, 1].reshape(X1.shape), alpha=.9)
plt.scatter(X_set[:, 0], X_set[:, 1], c=y_set, edgecolors="black")
plt.title('SVM (Training set)')
plt.legend((plt.Line2D((0,1),(0,0), marker='o', color=matplotlib.cm.get_cmap()(0.), markeredgecolor='black', linestyle=''),
            plt.Line2D((0,1),(0,0), marker='o', color=matplotlib.cm.get_cmap()(1.), markeredgecolor='black', linestyle='')), 
           ("No", "Yes"))
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.show()
```
